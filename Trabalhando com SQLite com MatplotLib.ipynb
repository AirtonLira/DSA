```python
import sqlite3
import datetime
import time
import random
import matplotlib.pyplot as plt
%matplotlib notebook

conn = sqlite3.connect('dsa.db')

cursor = conn.cursor()

def create_table():
    cursor.execute("""CREATE TABLE IF NOT EXISTS produtos(id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, date TEXT,
                  prod_name TEXT,valor REAL)""")
    
def data_insert():
    cursor.execute("INSERT INTO Produtos VALUES(10, '2019-05-01 14:00:00', 'teste', 90)")
    conn.commit()
    cursor.close()
    conn.close()
    
def data_insert_var():
    new_date = datetime.datetime.now()
    new_prod_name = 'Monitor'
    new_valor = random.randrange(50,100)
    cursor.execute("INSERT INTO Produtos (date, prod_name, valor) VALUES (?, ?, ?)",(new_date, new_prod_name, new_valor))
    conn.commit()
    
def dados_grafico():
    cursor.execute("SELECT id, valor FROM Produtos WHERE id <= 15")
    ids = []
    valores = []
    dados = cursor.fetchall()
    for linha in dados:
        ids.append(linha[0])
        valores.append(linha[1])
        
    plt.bar(ids, valores)
    plt.show()
```

```python
create_table()
for i in range(20):
    data_insert_var()
```
